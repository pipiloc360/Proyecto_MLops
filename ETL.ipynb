```python
#Vemos que los dataframes contienen informaci√≥n anidada dentro de algunas de sus columnas, por lo que se procede a sacar la informaci√≥n de los
#diccionarios
#Emepzamos a trabajar con la columna belongs_to_collection
#Esto es una funci√≥n que hace es aplicar literal_eval a un string para volverlo un diccionario y si se encuentra un vac√≠o 
#lo vuelve un None
def parse_dict_string(value):
    if pd.isna(value):
        return None
    else:
        return ast.literal_eval(value)
```

```python
#Vemos que de igual manera que con el primer archivo tenemos anidadas dos columnas
#Empezamos por la columna llamada cast
#aplicamos la funci√≥n a la columna cast que tambi√©n se encuentra cast
df2['cast'] = df2['cast'].apply(parse_dict_string)
```

```python
df_cast = pd.json_normalize(df2['cast']).fillna('')
```

```python
df_cast.head()
```

```python
def extract_actors(row):
    actors = []
    for col, value in row.items():
        if isinstance(value, dict):
            actor_name = value.get('name')
            if actor_name:
                actors.append(actor_name)
    return actors
```

```python
df_cast['Actors'] = df2.apply(extract_actors, axis=1)
df_cast.head()
```

```python
#ac√° separamos los diccionarios de cada columna en solo columnas y como solo necesitamos los nombres de los actores usamos ["name"] en el json_normalize
df_cast_desanidado = pd.DataFrame()

for column in df_cast.columns:
    desanidado = pd.json_normalize(df_cast[column])["name"].fillna('')
    df_cast_desanidado = pd.concat([df_cast_desanidado, desanidado], axis=1)
```

```python
df_cast_desanidado.head()
```

```python
#Creamos una copia por si algo sale mal, debido a que tenemos muchas columnas con nombres repetidos
df_cast_desanidadocopy = df_cast_desanidado.copy()
```

```python
df_cast_desanidadocopy.info()
```

```python
#Como muchas columnas quedan con el mismo nombre name, procedemos a enumerarlo para que cada columna sea independiente
cols = []
count = 1
for column in df_cast_desanidadocopy.columns:
    if column == 'name':
        cols.append(f'name_{count}')
        count+=1
        continue
    cols.append(column)
df_cast_desanidadocopy.columns = cols
```

```python
df_cast_desanidadocopy.head()
```

```python
#ahora lo que se hace es ya teniendo cada columna con nombre diferente, se itera sobre las filas para tener una serie por cada fila
#y luego los valores de cada columna de la serie se agregan a una lista, por lo que queda una lista de lista, donde cada lista es una de las 
#filas del data frame
actors = []
for index, row in df_cast_desanidadocopy.iterrows():
    fila = row[row != ""].values
    actors.append(fila)
        
```

```python
#Volvemos dataframe la lista
df_cast_final = pd.DataFrame({"actors": actors})
```

```python
df_cast_final.head()
```

```python
#concatenamos este dataframe con el original y eliminamos la colmna cast
df2= pd.concat([df2.drop('cast', axis=1), df_cast_final], axis=1)
```

```python
df2.head()
```

```python
df2.info()
```

```python
#Ahora hacemos lo mismo con la columna crew
df2['crew'] = df2['crew'].apply(parse_dict_string)
```

```python
#aplicamos el m√©todo json_normalize para que cada diccionario dentro de la lista quede en una columna
df_crew = pd.json_normalize(df2['crew']).fillna('')
```

```python
df_crew.info()
```

```python
#ac√° separamos los diccionarios de cada columna en solo columnas
df_crew_desanidado = pd.DataFrame()

for column in df_crew.columns:
    desanidado = pd.json_normalize(df_crew[column])[["job","name"]].fillna('')
    df_crew_desanidado = pd.concat([df_crew_desanidado, desanidado], axis=1)
```

```python
df_crew_desanidado.head()
```

```python
cols = []
count = 1
for column in df_crew_desanidado.columns:
    if column == 'job':
        cols.append(f'job_{count}')
        count+=1
        continue
    cols.append(column)
df_crew_desanidado.columns = cols
```

```python
cols = []
count = 1
for column in df_crew_desanidado.columns:
    if column == 'name':
        cols.append(f'name_{count}')
        count+=1
        continue
    cols.append(column)
df_crew_desanidado.columns = cols
```

```python
df_crew_desanidado.head()
```

```python
#Verificamos si existen m√°s de 1 director por pel√≠cula
filas_repetidas = df_crew_desanidado[df_crew_desanidado.apply(lambda row: sum(row == "Director") >= 2, axis=1)]
```

```python
filas_repetidas.head()
```

```python
#Ac√° lo que se hace es crear una lista de listas, en la cual cada lista representa una fila
#de igual manera se separa solo para job = Director
directores = []

for index, row in df_crew_desanidado.iterrows():
    directores_fila = []
    
    for i in range(1, 436):
        column_job = f"job_{i}"
        column_name = f"name_{i}"
        
        if row[column_job] == "Director":
            directores_fila.append(row[column_name])
    
    directores.append(directores_fila.copy())
```

```python
#finalmente se convierte a dataframe la lista
df_crew_final = pd.DataFrame({"directors": directores})
```

```python
#finalmente unimos con el dataframe original

df2 = pd.concat([df2.drop("crew", axis=1), df_crew_final], axis=1)
```
